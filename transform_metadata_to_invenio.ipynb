```python
def metadata(Name,MetadataID,Summary,OrganisationName,Information_owner_organization,ContactMailAddress,date):
    payload={
        "access": {
        "record": "public", ##   "restricted"
        "files": "public",  ##  "restricted"
        "status": "metadata-only"     ## or "open" ,   "restricted"
        },
        "embargo": {
            "active": False,
            "reason":None
        },
        "files": {
            "enabled": False, ## for metadata othrewise True
            "order":[]
        },
        "metadata": {
            "creators": [
                {
                    "person_or_org": {
                        "name": Information_owner_organization,
                        "type": "organizational"
                    },
                    "role":{
                        "id":"datacollector"}
                },
        
    # {
    #     "person_or_org": {
    #         "family_name": Information_owner_department_person.split()[1],
    #         "given_name": Information_owner_department_person.split()[0],
    #         "name": Information_owner_department_person,
    #         "type": "personal"
    #     },
    #     "role":{
    #     "id":"distributor"}
    #   },
        
    ],
            "description": Summary+",Information owner:"+Information_owner_organization+", "+ OrganisationName+",Contact:"+""+ContactMailAddress,
            "languages":{
                "id": "swe",
        },
            "publication_date": date,
            "related_identifiers": [
                {
                    "identifier": "https://dataportalen.stockholm.se/dataportalen/GetMetaDataById?id="+MetadataID,
                    "relation_type": {
                        "id": "isderivedfrom",
                    },
                    "resource_type": {
                        "id": "dataset",
                    },
                    "scheme": "url"
                }
            ],
            "resource_type": { "id": "dataset" },
            "title": Name,
        },
    }
    return(payload)
```
